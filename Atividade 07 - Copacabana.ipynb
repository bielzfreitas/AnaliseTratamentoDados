#### Exercício 01 - Faça um head() 

```python
copacabana = pd.read_csv('../data/Aula08_USJT_2020_GERDADOS-atividade7-copacabana.csv', delimiter=';')
copacabana.head()
```

#### Exercício 02 - Construa uma distribuição de frequência 

```python
posicao = [copacabana[copacabana['Posicao'] == 0]['Posicao'].count(),
         copacabana[copacabana['Posicao'] == 1]['Posicao'].count()
         ]
print(posicao)

tipologia = [copacabana[copacabana['Tipologia'] == 1]['Tipologia'].count(),
         copacabana[copacabana['Tipologia'] == 2]['Tipologia'].count(),
         copacabana[copacabana['Tipologia'] == 3]['Tipologia'].count()
         ]
print(tipologia)

fig = plt.figure(figsize=(20, 5))
gs = fig.add_gridspec(1, 2)

ax1=fig.add_subplot(gs[0,0])
ax1.set_title('Posicao')
sns.distplot(posicao, color="b")

ax2=fig.add_subplot(gs[0,1])
ax2.set_title('Tipologia')
sns.distplot(tipologia, color="r")
```

#### Exercício 03 B - Construa gráficos adequados

```python
# preço X distância da praia
df_graf1 = copacabana[['DistPraia','VAL_UNIT']]

# área construída X distância da praia
df_graf2 = copacabana[['DistPraia','AreaConstruida']]

fig = plt.figure(figsize=(20, 5))
gs = fig.add_gridspec(1, 2)

ax1=fig.add_subplot(gs[0,0])
ax1.set_title('Preço X Distância da Praia')
sns.distplot(df_graf1, color='r', kde_kws={"label": "DistPraia"}, hist_kws={"label":"VAL_UNIT"})

ax2=fig.add_subplot(gs[0,1])
ax2.set_title('Área Construída X Distância da Praia')
sns.distplot(df_graf2, color='r', kde_kws={"label": "DistPraia"}, hist_kws={"label":"AreaConstruida"})
```
